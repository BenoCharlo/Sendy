# **DATA PREPROCESSING**

## Data Separation

```python
target = train["Time from Pickup to Arrival"]
train.drop("Time from Pickup to Arrival", axis=1, inplace=True)
```

## Remove NA variables

```python
train.drop(["Precipitation in millimeters"], axis = 1, inplace=True)
```

## Variables to drop

```python
to_drop = ["Placement - Time", "Arrival at Destination - Time", "Confirmation - Time", "Arrival at Pickup - Time", "Vehicle Type"]
```

```python
train.drop(to_drop, axis=1, inplace=True)
```

## One_hot_encoding

```python
not_to_encoded = ["Distance (KM)", "Temperature", "Pickup Lat", "Pickup Long", "Destination Lat", "Destination Long", "Order No"]
```

```python
train_categorical = train.drop(not_to_encoded, axis=1)
```

```python
train_categorical = utils.MultiColumnLabelEncoder().fit_transform(train_categorical)
```

```python
train = pd.concat([train.filter(not_to_encoded, axis=1), train_categorical], axis=1)
```
